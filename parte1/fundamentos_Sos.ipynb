## Sistemas Operativos

Porque necesitamos un SO? Hay varios problemas que las computadoras enfrentan, y algunos de ellos son resueltos por los sistemas operativos. El primero de esos problemas es que las computadoras modernas son muy complejas principalmente en el sentido de cantidad de personas involucradas en el disenio y el armado. Tenemos hardware que es de INTEL y AMD, pero despues tenemos distintos fabricantes de mother que algunos utilizan hardware de INTEL o AMD o de terceros, a esto hay que sumarle que el usuario le pone una memoria de una determinada marca, etc. Entonces realmente las combinaciones de todos estos dispositivos son infinitas. Obviamente siempre esta el tema economico, todo fabricante que hace un programa quiere que su programa funcione en la mayor cantidad de maquinas posibles porque cada una de esas maquinas es un potencial cliente. Asi y todo el trabajo seria muy grande, cuantos modelos de teclado, de monitor de placas de video existen? muchisimas, fabricadas todas por distintos fabricantes. Aunque nos limites a un conjunto chiquito de dispositivos ya con estos dispositivos tendriamos miles de combinaciones.

## Programacion sin SO

Que es lo que deberiamos meter en un programa que hicieramos si quisieramos soportar por ejemplo un teclado, un monitor, una placa de red y un disco? Para el teclado por ejemplo deberiamos poner rutinas de manejo de teclado. Los scancodes del teclado no necesariamente estan standarizados. A esto deberiamos sumarle que el mismo fabricante con el mismo hardware hace un monton de distribuciones distintas, no es lo mismo el teclado en castellano que en ingles, etc. Hay que contemplar todas esas cosas. En cuanto a los monitores tambien necesitariamos rutinas de manejo de pantallas. Hay numerosas formas de mostrar video, distintas resoluciones de pantalla por ej. Lo mismo pasa con las placas y los discos. A todo esto se le agrega otra complicacion mas, como compartir las cosas que nisiquiera se quienes son. Yo hice un programa otro hizo otro programa como haces para no choquen. El problema entonces es como vamos a compartir los recursos. El otro problema es la seguridad entre tareas.

## Sistema Operativo

Para resolver estos problemas es que entra el sistema operativo. El SO divide todas estas cosas en varias capas. El SO claramente va a esta en el medio y va a implementar todas las funciones que son hardware independiente. Hay que dividir dos interfaces claras entre SO y driver y SO y aplicacion. El SO va a definir dos interfaces claras, una entre el SO y los dirvers y otros entre los SOs y la aplicacion. En los primeros el objetivo es claro. Por ejemplo los teclados, el teclado tiene que mandar una lista de scancodes, no importa como implementas tu hardware pero necesita una lista de scancodes estandar. Ahora no es que un programador tiene que ponerse a soportar todos los teclados del mundo, al programador le va a llegar la letra directamente con los acentos puestos por ejemplo, cada fabricante de un dispositivo tiene que encargarse de distribuir junto con el dispositivo un driver que es como una conversion de lenguaje entre el lenguaje implementado en el hardware con los chips que puso y un lenguaje estandar, entonces al SO el driver la a se la manda siempre con el mismo scancode. Esto obviamente no solamente ordena sino que divide mucho el trabajo y evita replicar el trabajo. El fabricante de un dispositivo tiene que hacer su driver para convertir su propio lenguaje en lo estandar. Entonces, por un lado el SO se queda en el medio con las cosas que no son ni hardware dependiente ni aplicacion dependiente, el que hace esta aplicacion no tenga que preocuparse de hacer las rutinas, entonces de esta forma distribuimos el trabajo y evitamos repetecion, cada uno no tiene que implementar su TCP esto hace se que el desarrollo de las aplicaciones sea mas corto y tenga menos bugs porque uso menos codigo. Ademas de esto, el SO lo que va a tratar de hacer es correr cada aplicacion dentro de una boveda, de forma que no interfieran con otros y si quieren hacer cosas fuera de las que tienen permitidas eso no pueda lograrse. Esta division es la que ha hecho posible el crecimiento del software. Inclusive para el usuario final es mas sencillo.

## SO moderno

Un SO para cumplir su fincion debe definir una interfaz clara para implementar sus drivers. Definir una interfaz clara a traves de la que expone su funcionalidad a las tareas que ejecutara, cualquier persona que quiera hacer una aplicacion pueda hablarle al SO. El SO virtualizara los dispositivos. Cada tarea piensa que tiene el teclado, monitor, interfaz de red, disco, etc para su exclusivo uso. Pero en realidad no tiene eso, lo tiene compartido. Como se comparten estos dispositivos? depende que dispositivo. Implementa mecanismos para que una tarea no espie ni altere otra. Si bien el programa que ve el usuario al arrancar el SO (sea en modo texto TUI o grafico GUI) no es parte del mismo sino que es una tarea mas, suele venir con el SO. Su funcion es abstraer el usuario del hardware tanto como sea posible y darle un manejo simple e intuitivo del sistema. Reparte la memoria entre las tareasd e forma que cada uno vea solo la suya.

## Un poco de historia

Al principio esto era medio inutil porque los primeros ordenadores que habian eran grupos de personas o empresas que hacian un hardware y luego lo operaban, inclusive a veces habian programas que requerian distintas configuraciones de hardware. En ese caso no hacia falta un SO porque era un grupo cerrado con un hardware unico. Cuando empiezan a comercialicarse se da el primero hito que destruye esto que es que los programadores se separab de los desarroladores de hardware. La forma que inicialmente se resolvio era tener un cjto de dispositivos minimo en el cual se ponian de acuerdo. Normalmente estos tenian solo dos dispositivos de E/S, un lector de tarjetas perforadas donde se ponia el codigo y una impresora donde se ponia el resultado. La cinta magentica fue el primer dispositivo magnetico para almacenar cosas y obviamente era mucho mas grande en capacidad que las tarjetas. El hecho del ingreso de la cinta magnetica mejoro mucho las cosas. Este primer programa que empezo a leer las cintas cargar la memoria, y encargarse de los resultados de devolerselos a la personas de alguna forma fue el primer SO, fue el primer programa que era comun a todas las tareas y que organizaba de alguna forma lo que pasaban en todas las maquinas. Mas adelante, con la expnasion de los calculos matematicos obviamente se perdia mucho tiempo entre una tarea y otra. Alli nace la multitarea pero de una forma muy basica porque en realidad los programas no corrian simultaneamente, entonces lo que se hacia era tener un procesador solo, un conjunto de dispositivos de entrada y salida y varias memorias de programa distintas. Un programador podia cargar la tarea en una memoria mientras que el procesador ejecutaba en otra memoria. Cuando terminaba impartia las memorias y se ponia a sacar los datos para luego imprimirlos o bajarlos a cinta, mientras tanto el procesador ejecutaba con la otra memoria el siguiente programa. De esta forma el hardware se aprovechaba mucho mas. Esto tuvo una ventaja importante para los programadores. En la decada del 70 aparecen las microcomputadoras a salir a las casas y aparece el primer SO (CP/M) de microcomputadoras que domino el mercado durante anios. A principios de los 80 IBM saca la PC y obviamente buscaba un SO, IBM pensaba en un negocio mas grande, fueron a ver a los propietarios de CP/M pero a estos no les gusto el negocio, entonces IBM compra un SO que era prototipo de una universidad, lo modifica un poco y genera el DOS que fue el primer SO que salio en las maquinas. Este SO DOS era monotarea, no estaba orientado a que sea multitarea, y encima era muy basico en cuanto a los dispositivos que reconocia. Al no reconocerlos era una solucion parcial para los programadores, estos no se tenian que preocupar por los teclados porque estos estaban soportados por algun driver del SO. A mediados de los 80 se pasa a modo grafico, impulsada por la multitarea, Windows y MAC OS cambian a una interfaz grafica. Nace la multitarea en las microcomputadoras personales y el concepto de "user friendly", (tratar de abstraer al usuario lo mas posible de las particularidades del hardawre) conceptos que siguen hasta el dia de hoy. En 1996 sale el windows NT 4.0 que fue un hito en el sentido de la seguridad, es el primer SO que implementa proteccion entre tareas y que la memoria de cada tarea esta restringida a esa tarea. Por ultimo para mencionar lo mas importante es linux que se hace popular en las microcomputadoras recien durante los 90 y su uso ha crecido hasta el dia de hoy, principalmente por su robustez y licenciamiento gratuito. Por ultimo, durante los 2000 de alguna forma las computadoras tenian un tamanio tan pequenio que dejan de ser simplemente computadoras personales, aparecen las tablets, celulares, etc.

## Proteccion entre tareas

Es imposible pensar en proteccion entre tareas sin pensar en ayuda del hardware. Si no tenemos ayuda del hardware deberiamos agarrar el codigo, interpretarlo y ver si afecta a otra tarea. Conceptualmente lo que se define en todos los procesadores modernos son un conjunto de anillos o privilegios, se dice que el anillo 0, el que esta mas adentro es el que tiene mayor permisos o mayor privilegio, y a medida que nos vamos para afuera se puede ir restinguiendo que cosas puede hacer el codigo que ejectuan esos otros anillos. Hay programas que corren en ring 0 y otros con privilegio menor. En definitiva se determinara que cosas se pueden hacer y que no en cada anillo, si alguien hace algo que no puede hacer con el privilegio que tiene, escencialmente esa tarea se suspende (se la saca del procesador y se le transfiere el control al padre). En el ring 0 se pone el SO que es el codigo mas critico, es el policia de la PC. En el ring 1 se ponen los drivers (necesitan permisos como acceder a hardware). Ring 2 es de tareas privilegiadas (login, manipulacion de credenciales, etc). En el ring 3 se ejecutan las tareas comunes.

## Tipos de SOs

Los SOs hacen distintas cosas cuando estan orientados a distintos mercados. Los SOs para mainframes, grandes comoputadoras que atienden miles de terminales, la idea es atender una gran cantidad de terminales pero normalmente no son en tiempo real sino que son trabajos que se encolan una atras de la otra. Son como transacciones. Los SOs orientados a los servers estan muy optimziados a atender comandos por la red. Tiene una capacidad para manejar muchos usuarios pero no fisicamente a traves de terminales sino a pedidos que le llegan por la red. Los SOs comunes normalmente es al reves, se tiende a darle mas prioridad a lo que hago en el momento y menos prioridad a lo que tengo minimizado, los SOs para servers es al reves. Despues hay SOs que estan especificamente pensados para sistemas multiprocesadores. Estos se dividen en dos grandes grupos, los que hacen computacion en paralela (son muy especificos para algunas cosas), y los de multi-computadoras (varias computadoras unidas por backbone) o multiprocesadores. Hay otro caso mas que son los sistemas embebidos se usan en equipos fisicos, televisores, heladeras, etc. Tienen una capacidad de entrada y salida limitada, pero esto se borra cada vez mas. Y tambien estan los SOs de tiempo real, para aplicaciones que no puede demorar ciertas tareas, como control de trenes, algunos procesos industriales, control de vuelo de un misil, etc. Usar un "tiempo virtual" puede terminar en grandes desastres. Algunos garantizan el cumplimiento de ciertas metas (hard) otros simplemente lo intentan (soft).
