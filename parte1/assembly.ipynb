## Lenguaje Assembler

Al dia de hoy todavia existen 4 grandes capas en los procesadores. La primer capa llamada modo real introducido en el 8086 original, era un modo que permitia correr aplicaciones de hasta 1 mega (cantidad total de direccionamiento del procesador). No habia ningun mecanismo de proteccion entre tareas. En el intel 80286 se introduce el modo protegido que es una proteccion de tareas que hace que una tarea no pueda acceder a la memoria del otro. Obviamente en este modo protegido no se podian usar ninguno de los beneficios que tenia el 286. El 386 sale con el objetivo de recuperar el mercado que habia perdido el 286. Tenia el modo virtual que permitia correr tareas sin proteccion dentro del modo protegido. Dentro del modo protegido armaba como una burbuja que se comportaba con el modo real. Lo mas importante es que permitia aprovechar la mayor cantidad de memoria. Luego AMD con el Opetron introduce el primer procesador de 64 bits, este processdor tiene un nuevo modo en el cual el sistema trabaja con todos registros de 64 bits en vez de 32. Estas son las 4 grandes capas que corren a dia de hoy.

Registros:

A: acumulador, normalmente esta pensado como registro de uso general

B: base

c: contador, cuando uno itera

dx: para levantar y guardar datos.

Hay dos grandes convenciones para hacer un ensamblador. La primera que arranco usando AT&T. Las particularidades son que los registros se los precede con un % y los valores inmediatos se los prececede con un $. En este modo se copia de izquierda a derecha. La otra es la convencion que usa INTEL, los parametros van de derecha a izquierda, en segundo lugar los registros no tienen un % adelante y en ultimo lugar las posiciones de memoria irian con corchetes. 

## Instrucciones

Operaciones con registros:

MOV  RAX, 0x200
MOV  RBX, RAX
MOV  RBP, 0x12000
MOV  QWORD PTR [RBP], RBX (muevo 4 palabras, 8 bytes, 64 bits)

Operaciones matematicas fundamentals

ADD  RAX, 0x12
MOV  RBP, 0x11000
SUB  DWORD PTR [RBP], 0x100
MOV  RAX, 0x6
MOV  RCX, 0x4
MUL  RCX
DIV  RCX

Operaciones matemticas fundamentales con enteros:

MOV  RAX, 0x6
MOV  RCX, -0x4
IMUL RCX
IDIV RCX

Operaciones con bits:

AND  RAX, RBX
OR   RAX
MOV  RBP, 0x100
XOR  AX, WORD PTR [RBP]
NOT  RAX

Rotaciones y desplazamiento de bits:

MOV  AX, 0b10100101
MOV  BX, 0b00010100
SHL  AX, 1
RCL  BX
SHR  EAX
SAR  EAX
ROR  EAX

El 8086 no tenia manejo de numeros de punto flotante.
