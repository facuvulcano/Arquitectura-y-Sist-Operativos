## Almacenamiento secundario

Un disco no es mas que una memoria en definitiva, en realidad uno parte de la memoria mas rapida que esta pegada al procesador que es el cache nivel 1, la segunda mas grande y mas lenta es el cache nivel 2, luego esta la RAM, y luego podemos ir a los discos y por ultimo a las cintas magneticas que normalmente son dispositivos que se usan para hacer back ups de grandes cantidades de informacion y que el acceso a esos elementos es muchisimo mas lento. Cuando hablamos de caches o la memoria hablamos de algunos nanosegundo mas o menos pero cuando hablamos de cintas hablamos de conseguir las cintas. En definitiva, si bien los dispositivos de almancenamiento masivo se los conoce como almanceamiento secundario la verdad es que es dificil trazar una linea de donde termina el primario y donde empieza el secundario. Hay tantos escalones intermedios es dificil ver donde esta essta separacion.

## Discos magneticos

Son los que mayormente manejan el mercado en cuanto a volumen de informacion almacenada. Los discos magneticos son un conjunto de platos stackeados unos arriba del otro cubiertos por un material ferromagnetico, y cada uno de estos discos tiene una o mas cabezas, generalmnete dos una va por encima y otra toca por debajo del disco con el objetivo de aprovechar las dos caras. Una cabeza es un electroiman, que producen campos magnetivos al aplicarles una corriente electrica (para grabacion) y generan una corriente electrica al ser expuestas al un campo magentico (lectura). Los discos giran y las cabezas se mueven en saltos discretos, es decir no se mueven en un continuo. La cabeza graba cilindros y va moviendose para adentro. Esto hace que deescriban trayectorias cilindricas a lo largo de la superficie magnetica. Dentro del cilindro, tampoco se guarda la informacion toda junta en un cilindro, no es un gran bloque de datos enorme sino que esta partido en pedacitos llamados sectores que es lo que se puede leer o escribir. Lo que se conoce como geometria del disco es precisamente saber cuanto cilindros tiene, cuantas cabezas tiene y la cantidad de sectores que hay dentro de cada cilindro. Esta geometria que se encontrara en los equipos como CHS, es lo que de alguna forma define como es ese disco, cada disco tiene una geometria distinta. Por estandar durante mucho tiempo los sectores tuvieron 512 bytes, hoy en dia todos siguen soportando 512 bytes, hay algunos mas modernos de 4096. Ese sector es la minima cantidad de informacion que se puede leer o grabar del disco. Al disco no le puedo decir leeme el tercer byte del sector tal, entendiendo al sector tal como cabeza tal, cilindro tal y numero de sector tal. A lo largo de la historia fue aumentando la cantidad de sectores que puedo meter en el disco, teniendo en cuenta que los sectores han permanecido del mismo tamanio lo que ha aumentado es la cantidad de sectores. Esto se hizo de distintas formas, una alternativa es meter cada vez mas platos, mas discos uno arriba del otro, esto aumenta el costo pero la realidad es que es un aumento lineal de la velocidad. Otra alternativa fue empezar a bajar el lugar o area que ocupaba cada sector, para eso habia que tener cabezas mas chiquitas que permiten leer y escribir en lugares mas chicos. En este caso tenemos un limite tecnologico, esto fue mejorando con el tiempo. Llega un momento que con este aumento de la densidad veo que los sectores de afuera son mas grandes, con la misma tecnologia podria grabar mas cosas. Los discos hoy en dia tienen distintas zonas, la zona de adentro tiene menos sectores que afuera. Los primeros discos exponian su geometria al SO, y esta ultima tecnica hubiera sido un problema. Sin embargo hoy los discos numeran los sectores en forma correlativa en lo que se ha llamado LBA. Cada disco tiene capacidad de computo propio que transforma este numero secuencial en los tres numeros CHS de acuerdo a la geometria de cada zona del disco. Asi y todo se empieza a llegar a un limite que no es tecnologico sino que fisico. La forma que vimos de guardar los datos en disco se conoce como "conventional magnetic recording" o CMR. Esta tecnologia se topo con un limite fisico que impidio seguir aumentando la capacidad. Cuando se trata de grabar cilindros muy delgados, las lineas magneticas van directo de un polo del iman al otro sin bajar lo suficiente como para orientar los dipolos magneticos del material del disco y grabar la informacion. Como no se pueden generar cilindros muy angostos, se opto por escribir superpuesto. Esta tecnologia se conoce como "Shingled Magnetic Recording" o SMR. Sin embargo esto tiene una clara desventaja: si se deseara modificar el sector rojo del grafico, no se podria escribirlo solo. Habria que escribirlo y luego reescribir el verde y luego el azul. O sea re-escribir toda la zona. Si bien el procesador dentro del disco resuelve esto en forma transparente, la perdida de tiempos es inevitable. Esots discos son mucho mas lentos que los CMR en estos casos.

## Discos opticos

Otra tecnologia muy utilizada hasta hace poco fueron los discos opticos, cds, dvds, blueray, etc. Escencialmnete lo que usaban era un sensor de luz para medir la reflexion de un haz de luz generado con un laser en una superficie reflectiva. Cuando el laser se usaba a baja potencia se usaba para que rebote contra el disco y medir la reflexion, cuando se usaba en alta potencia calentaba la superficie reflexiva, daniandola haciendo que no refleje mas, de esa forma se podia grabar informacion en los discos. La forma de grabacion era quemar ese material reflectivo generando puntos y rayas y de esa forma se entendian como ceros y unos y se grababa asi la informacion. El problema era que la superficie degradada no se podia volver a hacer reflectiva facilmente. Peor aun los procesos de borrado que se encontraron no podian borrar solamente un sector pequenio. El tener que pasar por un proceso de borrado total, unido a los errores generados por rayones, y a que las memorias alcanzaran densidades de grabacion similares, hicieron que se dejaran de usar.

## Discos de estado solido

Hoy en dia la mayor parte de los discos de alta densidad estan hechos con memorias del tipo flash. Escencialmente hay dos tecnologias hoy en dia, tecnologia nand or nor. 



Las memorias NAND tienen mayor densidad de almacenamiento se usan mucho para los discos SSD donde uno tiene grandes volumenes de datos. Obviamente requieren del ECCs para que funcionen. Las NOR se han usado batante en celulares, permiten acceder byte por byte (no de a paginas), por lo que se usan para conectarlas en lugar de ROM a procesadores, como en el caso de los ceulares. Son mas rapidas, mas lentas para algunas cosas pero es un tema costo beneficio. La realidad es que las dos tienen el tema de los bloques, que hay que leer sectores o zonas.

## Estructuras logicas dentro del disco

Un disco en cualquiera de sus variantes, desde discos magneticos, flash, etc.. en definitiva es un conjunto de sectores de tamanio fijo todos tirados ahi en el medio sin ningun tipo de organizacion logica. El hardware me los da todo suelto. Estos sectores simplemente estan enumerados de acuerdo al estandar LDA. Como convertimos ese conjunto de sectores sueltos en una unidad utilizable por los programas que tenga archivos, directorios, permisos, etc. Como usuario deseariamos que el disco tenga una forma de detectarlo como tal, si conecto un disco a un sistema, el sistema pueda decir este disco que tiene. Otra cosa que ahora tal vez es menos usada es que a veces un disco no se le quiere ver como un unico disco sino como discos mas pequenios, lo que se llama particionar el disco. Otra cosa que necesitariamos es arrancar el SO que va a estar en disco, como podemos hacer para generar algun tipo de norma que todos puedan seguir para poder alcanzar estos objetivos, y que cualquier bios pueda detectar cualquier disco independientemente del fabricante. Esto se ha resuelto hace muchos anios, se resolvio escribiendo un bloque especial en el primer sector del disco. Ahi se tenia el master boot record que tenia una tabla de particiones, que tubo algunas limitaciones y temrino evolucionando en un sistema de particionado nuevo de indentificadores unicos o GUID. El MBR es el mas viejo.

## Master Boot Record

Que tenia ese primer sector de todos los discos? Lo primero que hacia era analizar los ultimos 2 bytes para ver si valian 0x55 0xAA, era como una marca. Si valian eso entonces asumia que era una particion valida y analizaba el resto del contenido. Si la marca estaba bien habia entoces un arreglo de cuatro estructuras de cada parte del disco y un identificador unico de disco. El identificador de disco se generaba al azar. La tabla de particiones era el vector de cuatro cosas, la idea era tener el disco partido en distintos discos logicos, cada una de estas particiones es un vector de 4 elementos de estas estructuras. Tenia 1 bit que era si estaba activo, es decir si esta particion estaba declarada o no declarada. 7 bits libres, 24 bits del CHS del primer sector. 8 bits de tipo de particion, 24 bits para el CHS del ultimo sector. Todas las combinaciones del CHS nos daban 2^24 sectores que daba 8GB al multiplicar por 512, luego el CHS se cambia a LBA, que limita a 2^32 * 512 = 2 Tb. Seguimos con el proceso de arranque de la maquina, cargaba del primer sector, se fijaba la firma, validaba que sea 55 AA, esto quiere decir que la particion tiene informacion util, leia los 4 elementos del arreglo buscando alguno que estuviera activo, cuando encontraba uno activo decia que esa era la paarticion que hay que hacer. En el pedacito que tenia 440 bytes se le dejaba al fabricante al SO que ponga un programa. Este programa se encargaba de levantar el SO. ese programita lo que hacia era levantar otros sectores que estaban en lugares especificos y obviamente ese programa iba creciendo y cargando rutinas mas complejas. Pero la realidad es que ese proceso de arranque se hacia sin tener una estructura de archivos en el disco.
