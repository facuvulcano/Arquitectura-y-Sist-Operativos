## Buses

El bus conecta dos bloques que tienen que intercambiar informacion. Estrictamente son lineas electricas que tienen una funcion a fin, que es implementar esa comunicacion. Hay una separacion entre el buss de adress y bus de datos, todos los bits del bus de adress tenian una funcion en comun que era establecer que de direccion de memoria queremos leer o escribir, el bus de datos, al reves era para leer o escribir de la memoria. Entonces, en realidad un bus es un conjunto de lineas electricas que tienen una funcion afin, destinadas a intercambiar informacion entre dos bloques, cualquieras sean, de un ordenador. Cuando uno quiere conectar una memoria al procesador, eso esta todo metido en la placa base y es muy predecible para el que disenia. El problema aparece cuando queremos conectar dispositivos externos, el que disenia el mother no sabe que es lo que van a conectar, la gracia es que sea lo mas abierto posible para que este sujeto a conectarse a la mayor cantidad de dispositivos externos posibles. El bus nos dara la conectividad electrica y la posibilidad de conectar 2 o mas extremos. Habra que escribir el driver correspondiente a los efectos de que se va a escribir y como comunicarselo al SO.

## Tipos

Si nos basamos en la arquitectura, principalmente en el grado de paralelismo podemos dividir en dos grandes grupos. Los paralelos: la informacion viaja por varias lineas en formas simultanea, por ejemplo el bus de address, lo mismo que con el bus de datos. Hay otros que son en serie, hay una linea de datos y la informacion se serializa mandando 1 bit a la vez uno atras del otro. Cada uno tiene sus ventajas y desventajas. Si pongo buses que son paralelos, como tengo muchas lineas para usar juntas el ancho de banda va a ser mayor. Uno puede sacrficar velocidad en pos de practicidad, con la practicidad viene la economia, asi son los series, es decir mas baratos pero con menor ancho de banda. 

## Velocidad

Hay parametros fisicos que limitan la velocidad a la que se pueden transimitir los datos. Entre los que mas influyen estan: El tamanio del bus: parece ridiculo cuando lo piensa a priori pero si pensamos en cuanto tarda la luz en barrer 35 cm vemos que da algunos nanosegundos, tenemos compuertas logicas que basculan en menos de nanosegundos, la limitacion esta entonces que la electricidad no llega. Otro factor importante es la tension, nivel de voltaje que se pone en los buses. En todo ambiente hay ruido electrico que se mete, y es ruido que se suma a las lineas, entonces si tengo tensiones muy chiquitas el ruido me puede confundir. Usar tensiones altas puede tener como desventajas que por un lado hay que resistir esas tensiones, y por otro lado hay un tema de consumo electrico, y la energia sube con el cuadrado de la tension, entonces usar tensiones mas altas implicara mayor consumo de energia. Durante muchos tiempo los procesdores usaban 5 volts y recien los procesasdores de las notebooks usaban 3.3 volts y ahora incluso utilizan menos, porque obviamente consumir menos en el caso de una notebook es sinonimo de mayor duracion de bateria, entonces hay un impacto mas directo en el usuario. Tambien depende si estoy usando multiples lineas para mandar un solo bit eso tambien influye en la velocidad. Siempre que un transmisor quiere enviar un bit al otro lado va a poner distitnos valores de potencial o de volts de un lado y del otro lado va a haber alguien que va a tener un valor de discriminante que diga de aca para arriba es un uno y de aca para abajo es un cero. El problema es que en el camino se suma ruido electrico, habra que discriminar eso como para que el ruido no me confunda y pueda garantizar que el receptor no se vera influenciado por el ruido. Hay una solucion inteligente que es utilizar dos cables para mandar que es lo que se llama informacion diferencial. Mando dos cables y le digo del otro lado no mires el potencial absoluto sino que mira la diferencia de potencial entre los dos cables, estos cables se suelen trenzar de forma que no haya ninguno mas cerca o mas lejos del ruido, yo se que en el medio hay ruido pero ambos seran expuestos al mismo ruido entonces finalmente el ruido se restara y cancelara. De esta forma bajaremos los potenciales y el consumo. Hay buses que son diferenciales y buses que no, normalmente los que estan dentro del mother (como el bus de adress, datos) esos no suelen ser diferenciales, para equipos que van conectados afuera, como el bus usb o hdmi esos si, estan mas propensos a agarrar ruido. Como el tiempo de propagacion de senial es un factor, la realidad que todas las lineas tardan en llegar al otro lado y eso tiene que conisderarse en el disenio. El gran avance de las frecuencias de clock fue donde practicamente se multipclaba por 10 cada decada, es decir ese gran aumento de velocidad se dio por la miniatorizacion. Esa mayor velocidad tambien involucra mayor disipacion de energia, hoy en dia hay brutos disipadores en los procesadores, hoy en dia ya no hay mucho marguen para achicar. 

## BUS ISA

Este bus ya no se usa estrictamente, servia para conectar distintas placas a las computadoras, hoy tambien se hace con el PCI. del adress 0 hasta el 19 son los bits de bus de adress, data 0 hasta data 7 son los bits del bus de datos, tambien teniamos las lineas de interrupcion, hay una direccion de read y una de write y tambien estan los canales de DMA. Estos buses estan muy claramente documentados y cada diseniador podria hacer lo que quiera. Soportaba transferencias de 8 y 16 bits a la vez, y habia 8 mega transferencias por segundo con 2 waits states (se usan cuando quieren acceder a un medio mas lento). La velcoidad de transferencia era de 5 mbps.

## BUS PCI

Es el sucesor de los ISA, el PSI empieza a principios de los 90s y lo primero que hicieron fue subir a 32 bits de datos. Poner mas bits implica que esten mas pegados por ende requieren mayor tecnologia para hacer los conectores y para hacer los contactos que despues van a entrar ahi. Se baja la tension a 3.3 volts con la idea de bajar el consumo, permitiendo llevarlo de 8 mega transferencia por segundo a 33 mega transferencias por segundo. Soportaba transferencia opcionalmente hasta 64 bits, cuando uno aumenta la cantidad de cables en el conector, por un lado duplican los bits de datos, con lo cual cada transferencia tenia el doble, pero aparte aumentaron la cantidad de bits de adress. El bus PSI sube de 5 mbps a 133 mbps.

## BUS PCI Express

Los conectores que tienen hoy las placas de cualquier PC son los PCI Express, de los cuales tenemos una familia de conectores. Porque tenemos conectores chiquitos y conectores mas grandes? porque tenemos una presion doble, por un lado la presion de las placas de video que cada vez quieren tener mas bits de adress, por otro lado estan los fabricantes de otros dispositivos que quieren pocos bits de adress. Frente a esta presion mutua lo que se hizo fue crear distintas alternativas para quedar bien con todos. Primero, para poder aumentar la velocidad y dado el problema del ruido ahora los canales son diferenciales. Hay distinta cantidad de pares de cables, cada par se llama un link, y la velocidad de transferencia de ese link es de un giga byte por segundo. En que difieren los distintos conectores? tienen distinta cantidad de links entonces el PCI Express 1 tiene un link solo osea que esta limitado a un giga byte, pero el mas grande, tiene 16 links, obviamnete una placa de video probablemnte utilize uno de 16 links. Se lleva 16 bytes por segundo pero esto a ido aumentando con distintas revisiones para poder soportar mas velocidad. Ya vamos por la version 6 del PSI Express con una tasa de transferencia de 64 GBs en cada direccion, o sea que pueden leer en 64 y escribiendo en  64. Recordaque que arrancamos en 5 mbps y ahora estamos 126.000 mbps. Hoy en dia este tipo de conectores es el standard en todas las PCs.

## Hot Swapping

Los distintos buses uno puede tenerlos en una placa madre en un conector y en realidad es esperable en ese caso que la persona que va a conectar una placa nueva apague la maquina, desenerginze todo, conecte la placa y despues arranque la maquina denuevo, eso permite una electronica mas sencilla. No es muy practico tener que desenchufar la maquina para conectar un pendrive por ejemplo. Hoy en dia la mayor parte de los teclados por ejemplo son USB. Entonces existe esa diferencia entre si los buses soportan la conexion en caliente, es decir con el sistema energizado y funcionando y con los que no lo soportan. Esa diferencia es dual, tanto de software como de hardware. Por el lado del hardware, hay que preveer que el usuario pueda enchufar distintas cosas, y por otro lado hay que modificar el software, si yo agrego un dispositivo hay que cargar un driver para que pueda ser utilizado. Entonces para los buses que soportan conexion en caliente necesitamos hardware apropiado (que no se queme cuando conectamos cosas con energia) y por otro lado necesitamos software que sea capaz de detectar esos cambios. 

## BUS SATA

El primer bus que soporta hot swap es el actual bus que se usa para los discos rigidos. El SATA se disenio especificamente para transferencia de datos entre almacneamiento masivo, para conectar discos. Originalmente tenian una tasa de 600 mbps, de hecho es la actual. No son cables muy largos, el mas largo es de 2m. Hoy se usa tanto como para notebooks como para maquinas de escritorio. La norma es que soporta hot swapping, osea que soporta desenchufar el disco en caliente, esto es muy importante cuando hablamos de RAIDS.

## M2

No todos usan SATA, cada vez son mas comunes por el uso de los SSD. Tienen una veloicdad de transferencia similar a SATA. Los que tienen sufijo B utilizan un solo link mientras que los que tienen sufijo M utilizan 4 links. Algunos son SATA, y utilizan el puerto SATA disponible en el conector donde se los inserta.

## USB

USB nace como un concepto totalmente diferente a lo que eran los buses de transferencia masiva para discos. No era lo principal la velocidad, el objetivo principal era que sea versatil, es decir que se pudiesen coectar multiples cosas, y ademas que sea barato. Entonces, la condicion era hacer algo rapido y versatil pero con algo que en su momento era algo bastante innovador que era el plug and play. Hasta ahora habiamos hablado de normas de conexion de hardware pero cuando sacan el USB, por primera vez el sistema reconoce que le conectaron y carga el driver solo. Hace no mucho tiempo cuando uno compraba el dispositivo tenia que conseguir el driver. La condicion de disenio del USB era decir porque cuando un dispositivo se conecta no manda de alguna forma quien es o que es y ponemos una base de datos de donde se descargara el driver automaticamente. Entonces las condiciones de disenio eran, versatil, bajo costo, no abrir el gabinete para conectarlo, es decir tenia que ser para cualquiera y obviamnete que funcionara sin que el usuario haga nada. El USB 3.0 tiene del orden de 600mbs de velocidad de transferencia maxima. Con estas condiciones que se les ocurrio? El vinculo entre el bus y el ordenador es el dispositivo llamado root hub. Al mismo se pueden conectar dispositivos u otros hubs. Entonces tenemos dos tipos de dispositivos los hubs y los terminales. El root hub es el que tiene la cantidad de salidas que estaran conectadas en los conectores de los slots usb que tienen. Es totalmente master-slave. El procesaodr a traves de un root hub transmite comandos a traves de estas lineas, caundo un hub lo recibe lo que hace es replicarlo por todos sus cables que tenga conectado. Esos paquetes que manda el root hub pueden ser distitnos tipos. El primer paquete es el de discovery, como cada dispositivo tiene un numero unico, cuando un disp se enchufa asume que tiene 0, entonces al enchufarlo periodicamente viajan paquetes que se fijen si alguien tiene el 0. Lo normal es que nadie responda cuando el procesador manda el paquete, hasta que enchufe uno. Despues los otros paquetes, como cada uno tiene un numero unico puede haber paquetes para un dispositivo en particular que el driver quiere enviarle alguna informacion. Todos los paquetes van por dos lineas. El root hub transmite un comando, libera las lineas y espera que alguien le conteste o no le conteste nadie. Aproximadamente una vez cada milisegundo se inician esas transferencias. Es totalmente master-slave, siempre el root es el que transmite y espera que alguien le conteste o que no le conteste nadie. Los datos viajan en las dos direcciones por el mismo par de cables y que el cable es un bus diferencial porque permite cables largos. Cuando el SO recibe el paquete de discovery, obviamente tiene que saber el tipo de dispositvo para ver que tipo de driver va a cargar. Dentro de esa respuesta del paquete de discovery lo que envian es un byte con la clase de dispositivo. Para tratar de bajar el costo sobretodo de los que fabrican hardware se ha creado dos bytes mas que viajan dentro del paquete de discovery que son la subclase y un protocolo.
