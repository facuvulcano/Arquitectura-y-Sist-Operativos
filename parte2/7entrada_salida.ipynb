## Teclado

Arrancaron como tablas con teclas conectadas en una placa base, si bien los teclados son de las cosas que menos cambiaron, es decir del lado del usuario quizas cambiaron poco pero por debajo si han pasado cosas. Con el tiempo las teclas fueron migrando a alternativas mas economicas. Los teclados de membrana ya no tienen mas teclas sino que tienen los dos contactos dibujados en el impreso y hay una tecla arriba de una lamina de goma que tiene una parte conductora. Cuando se aprieta la tecla, la tecla baja y une los dos conductores. El teclado de membrana permitio bajar drasticamente los costos. Hoy en dia con mercados mucho mas especificos como los juegos han vuelto a aparecer muchos teclados con teclas fisicas inclusive con distitnas fuerzas de accion. Metiendonos debajo, en el funcionamiento interno, como dijimos antes, el teclado tenia algun chip controlador que lo que hacia era mirar que teclas estaban apretadas y reportarle al procesador, interrumpir al procesador y mandarle un numero de tecla cuando se apretaba o soltaba. Ese inicio tan basico del teclado se mantiene hasta hoy, todas las maquinas actuales interrumpen el procesador cuando se hace el contacto el procesador recibe un codigo de make, cuando uno la suelta se recibe un codigo de breake. Durante mucho tiempo la mayor parte de las teclas no permitian apretarse simultaneamente. Esto se debe a que internamente se disponian como una matriz, cada tecla puenteaba una fila con una columna, escencialmente el controlador del teclado ponia un 1 en solo una de las filas y censar que veia por las columnas o visceversa, si no tenia ninguna tecla apretada veia todo en 0, si alguien presionaba una llave subia a 1. Acto seguido, se pasaba a la siguiente poniendo un 1. Esta tecnica es muy economica, con poco hardware podia poner muchas teclas. El problema que tiene es que si aprieto mas de una a la vez puedo tener porblemas. Durante bastante tiempo las teclas que permitian la presion en paralelo eran las teclas de control. El mercado de los juegos impulso la necesidad de que n teclas se puedan apretar simultaneamente y puedan ser reportadas al procesador como tales, obviamente implican hardware y mayor costo.

## Ratones

Los mouse tambien siguieron una larga etapa de evolucion. Inicialmente eran una bolita que movian dos rodillos y giraban un potenciometro, a medida que iba girando cambiaba el valor de la resistencia. Obviamente tenian sus problemas porque estas resistencias iban y venian, y se terminaban gastando. Durante mucho tiempo se cambio los potenciometros por encoders. Actualmente unos LEDs emiten luz y una pequenia camara saca fotos unas 1000 veces por segundo. Un pequenio proceador analiza el movimiento entre cuadros de las imperfecciones sobre las que se desplaza el raton y determina la direccion del movimiento. Las fotos se comparan unas con otras para ver en que direccion se movio. Hoy en dia los ratones funcionan sobre muchas superficies salvo que la superficie sea muy uniforme. La longitud de onda que usen los ratones impacta en el funcionamiento en distinta superficies. El proceso de deteccion es a traves de la utilizacion de algoritmos. Por cada unidad de movimiento reportan una vez y envian al CPU cuanto se movio el dispositivo y el estadao de los botones.

## Pantalla Tactil

Nacieron de una forma muy basica, habia un conjunto de rayos infrarojos que emitian de un lado y un conjunto de receptores de otro. Este conjunto de haces eran de baja resolucion. Luego aparecieron los llamados resistivos, comunes en la decada del 90. Consisita de dos membranas transparentes, la superior fexlible con pequenios hilos conductores. Cuando uno presionaba doblaba internamente esa capa de plastico y hacia que uno de los conductores tocara con el otro. Esto aumento drasticamente la resolucion, tenemos el mismo problema de tocas en mas de un lugar al mismo tiempo pero en ese momento estaba pensado mas para botones y comandos. Hoy en dia se elimino el tema de usar las membranas plasticas, hoy en dia estan hecho por capas de vidrio con pocos atomos de espesor, el concepto es el mismo, tener un conjunto de filas y columnas pero hay dos cosas que cambiaron a nivel interno. Primero al usar vidrio, el vidrio no cede, por lo que no podemos decir que son resistivos, escencialmente se mide un fenomeno que es la capacidad, cuando uno acerca el dedo cambia la carga electrica y esa carga electrica es la que se detecta. Tambien se cambio la forma de los controladores, se paso a usar el mismo sistema que los teclados, hoy en dia soportan varios dedos al mismo tiempo.

Las primeras maquinas no tenian monitores, tampoco servian para mucho porque los programadores no estaban delante de la maquina, hasta que surgen las primeras terminales cosa que amaron mucho los programadores porque generaba un ida y vuelta que antes no existia. Los primeros monitroes eran tubos de rayos cadoticos, hay un emisor de electrones esos electrones se aceleran hacia adelante formando un rayo y cuando esos haces pegan contra el frente, devuelve del otro lado. Mientras alguien prendia y apagaba el haz el haz era movido por la bobina de forma que barra la fila. Cuando terminaba volvia y empezaba devuelta. La retencion de luz que tiene el ojo humano es lo que nos daba la percecpcion de una imagen completa cuando en realidad se dibujaba de un punto a la vez. Los monitores de cristal liquido tenian un funcioanmiento similar, tambien estan divididos en puntos, tienen un liquido especial metido entre dos vidrios con un polizador. La luz tiene posicion, tiene ondas en todas direcciones, uno puede hacer que pase por un determinado filtro y que pase la luz que tiene una determianda posicion, el tema es que hay materiales como el cristal liquido que bloquean la luz cuando pasa en cierta direccion. El concepto detras del monitor de cristal liquido es poner luz que viene solo en una direccion. El problema de los sistemas de cristal liquido es que el tiempo de alinear y desalienar las moleculas no se puede controlar con exactitud, cuando uno las quiere alinear con campo electrico hay una regla, el problema esta cuando remuevo el campo electrico. De hecho los monitores LCD hoy en dia no se utilizan, se ha pasado a los monitores LED. Hoy en dia la tecnologia actual son matrices de LEDs, cada punto tiene tres leds, azul, verde y rojo y graduando la intensidad de estos colores se puede reprsentar cualquier otro color. Los leds tardan mucho menos en prenderse y apagarse. 

## Impresion

Lo que habia cuando empiezan a salir las impresoras eran las maquinas de escribir y obviamnete compartian mucho con las maquinas de escribir las primeras impresoras. Escencialmente uno tenia el papel y una cinta adelante que tenia tinta, la maquina de escribir golpeaba la cinta con un tipo predefinido y eso dibujaba la letra en el papel, en el caso de las impresoras lo que tenian eran agujas que al empujarlas para adelante empujaban la cinta contra el papel y dibujaban un conjunto de puntos, las agujas se retraian, se avanzaban una fraccion de milimetros para el costado, se volvian a disparar, etc.. para ir dibujando toda la fila. Este conjunto de las agujas estaban una encima de la otra y eran movidas por electroimanes. La tecnologia se copio de las maquinas de escribir, es muy anterior a los monitores. Cuando se queria imprimir color no habia muchas alternativas pero tambien se copio de las maquinas de escribir. La siguiente tecnologia fue llamada chorro de tinta, esta en el mercado hace bastante, escencialmente tiene un cabezal que tiene una fila de agujeros que van a dibujar los puntos. Se pasa al sistema de chorro de tinta con el mismo sisteam de dibujar fila por fila y luego pasara al de abajo. Escencialmente tiene un pedazo de conductor con una resistencia, se coloca un dispositivo que tiene tinta y con un orificio abajo, ese orificio es tan chico que la tinta no llega a caer, la forma de forzar a que salga es calentar muy rapido la resistencia generando una burbuja de vapor en la tinta que se expande rapidamente y genera que salga un chorro de tinta hacia el papel. Simultaneamente uno tiene que tener un monton en paralelo para ir dibujando de a franjas. Como funcionan los que tienen cartuchos con multiples colores? en vez de tener una sola filita todas con el mismo color, tienen 3 (rojo, verde, azul) o 4 (rojo, verde, azul negro) filas separadas bastantes. Lamentablemente involucraba el cambio del cartucho completo, lo cual era muy caro. La ultima tecnologia de impresoras es la laser, a diferencia de la impresion vista hasta ahora, se iba construyendo de a poco por filas, las impresoras laser son un proceso que se da de forma continua a lo largo de un rodillo, mientras se esta haciendo una cosa se hace otra con lo que viene despues. El proceso arranca cuando un sistema de limpieza limpia ese rodillo. El rodillo tiene una pelicula con un material verdoso, ese cilindro lo que tiene es la particularidad de tener un material fotosensible en la superficie. Lo primero que sucede es que un sisteam de limpieza lo limpia asegurandose que arranque de cero y que no hay restos de lo que se imprimio antes. Todas las filitas que tiene la impresion son como pequenias lineas sobre ese cilindro que van pasando al papel. El proceso limpieza, carga, dibujo, deposito, transferencia y fusion es el proceso contiguo que se da. Limpieza se limpia el rodillo, luego pasa por un sistema de carga que carga negativamente el cilindro, acto seguido hay un laser que pasa de lado a lado del cilindro emitiendo pulsos de luz, en los lugares donde pulsa la luz calienta el material y hace que la carga que esta contenida se disipe, el material se hace conductor con la luz y la carga se va. Osea que al pasar por el lacer me quedan muchos puntos algunos con carga electrica y otros sin carga, acto seguido viene para el toner, el toner es un material con particulas muy chiquitas que fueron cargadas positivamente dentro del cartucho, cuando pasa cerca el rodillo, en los lugares donde tiene caraga negativa el toner se adhiere, en los lugares donde pego el laser y libero la carga negativa, el toner no se adhiere. Al ser expuesto al toner, la parte del rodillo con carga negativa atrae particulas de toner, que son transferidas al papel con carga positiva. Por ultimo un fusor calienta el polvo para convertirlo en una capa fija al papel. El proceso es continuo y simultaneo.

## Modems

Viene de modulador-demodulador. Modular una senial implica alterar alguno de sus paraemtros para agregarle informacion, si miramos la ecuacion matematica general de una senoidal seria $ V(t) = A * sen(\omega t + \phi)$. A es la amplitud omega la pulsacion angular y phi la fase inicial con que arranca. Se puede modular alterando cualquiera de esas variables. Hay distintas variables que se pueden usar para modular. El detector del otro lado tiene que ser lo suficientmente sensible como para detectar los cambios de la modulacion, del lado de la demodulacion es inverso. Estas seniales pueden transmitirse por distinos medios: Seniales de audio por una linea telefonica (Modem telefonico), seniales electromagenticas transmitidas por radio (placa WIFI o Bluethooth) seniales electromagenticas en un cable-modem o modem DSL. Pulsos de luz en un control remoto o fibra optica. Es un procedimiento general que permite transmitir informacion de un lugar a otro. La velocidad maxima de transmision de datos depende de la frecuencia de la senial que se modula, la cual varia en funcion de los medios usados para transmitirlas.
