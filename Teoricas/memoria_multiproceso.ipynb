## Ambiente multiproceso

En un ambiente multiproceso multiples procesos se ejecutan simultaneamente, estos procesos comparten recursos como la CPU, memoria, etc. Es crucial la gestion eficiente de la memoria para garantizar un rendimiento optimo, la seguridad y la estabilidad del sistema.

## 1. Espacio de Direccion de Memoria:

Cada proceso en un sistema multiproceso generalmente opera en su propip espacio de direcciones virtuales. Cada proceso tiene la ilusion de tener acceso exclusivo a la memoria, aislado de otros procesos. El SO utiliza uan unidad de gestion de memoria (MMU) para mapear estas direc virtuales a direc fisicas en la RAM.

## 2. Memoria Privada vs. Memoria Compartida:

- Memoria privada: Cada proceso tiene su propio espacio de memoria que no es accesible por otros procesos. Asegura aislamiento y seguridad.

- Memoria compartida: A veces, es necesario que varios procesos accedan a una misma region de memoria para comunicarse o compartir datos. Es una tecnica eficiente para la comunicacion entre procesos (IPC).

## 3. Sincronizacion y Consistencia:

Cuando multiples procesos acceden a la misma region de memoria compartida, es fundamental asegurar que no ocurran condiciones de carrera o incosistencias. Para ello se utilizan mecanismos de sincronizacion como:

- Mutexes (mutual exclusion): Evitan que multiples procesos accedan simultaneaamnte a una seccion critica de codigo o a una region de memoria.

- Semaforos: Controlan el acceso a recursos compartidos permitiendo un numero limitado de accesos concurrentes.

- Barreras: Sincronizan multiples procesos para que alcanzen ciertos puntos en la ejecucion antes de continuar.

## 4. Gestion de Memoria Virtual:

La memoria virtual permite que cada proceso vea un espacio de direcciones continuo y aislado, independientemente de la fragmentacion fisica de la RAM. Esto se logra mediante tecnicas como:

- Paginacion: Divide la memoria en bloques de tamanio fijo llamados paginas. Las paginas pueden estar en memoria fisica o en almacenamiento secundario (como el disco duro).

- Segmentacion: Divide la memoria en segmentos de tamanio variable, como codigo, datos y pila.

La memoria virtual facilita la multiprogramacion, permitiendo que multiples procesos residan en memoria simultaneamente sin interferir entre si.

## 5. Intercambio de paginas (Swapping):

En sistemas donde la memoria RAM es limitada, el sistema operativo puede intercambiar paginas de memoria entre la RAM y el almacenamkiento secundario para liberar espacio. Este proceso puede afectar el rendimiento si se realiza con frecuencia (thrashing).

## 6. Proteccion de Memoria:

El SO implementa mecanismos de proteccion de memoria para evitar que un proceso acceda o modifique la memoria de otro proceso sin autorizacion. Esto previene errores y mejora la seguridad del sistema. Algunas tecnicas:

- Tablas de paginas con bits de permiso: Indican si una pagina es de solo lectura, lectura/escritura, ejecurable, etc.

- Modo de usuario y modo kernel: Separan los privilegios, impidiendo que los procesos de usuarios realicen operaciones sensibles directamente.

## 7. Optimizacion y caches:

Para mejorar el rendimiento en ambientes mutliproceso, se utilizan caches de memoria y algoritmos de reemplazo de paginas eficientes que reducen la latencia de acceso a datos frecuentes. Ademas la localidad de referencia se aprovecha para optimizar el uso de la memoria.

## 8. Fragmentacion de Memoria:

La fragmentacion interna y externa son desafios en la gestion de memoria. Los sistemas multiproceso deben implementar estrategias para minimizar la fragmentacion como:

- Asignacion de memoria basada en bloques: Utilizan tamanios de bloques que reduzan el espacio desperdiciacdo.

- Compresion de memoria: Reorganizar la memoria para consolidar bloques libres y reducir la fragmentacion.

## Hay dos problemas fundamentales en la gestion de memoria en un ambiente multiproceso:

- Reubicacion de ejecutables

- Proteccion entre tareas

## 1. Reubicacion de ejecutables:

Se refiere al proceso de cargar un programa (ejecutable) en una ubicacion especifica de la memoria fisica durante su ejecucion. Dado que cada proceso puede requerir diferentes areas de memoria y que el espacio de direcciones virtuales es independiente para cada proceso, es esencial manejar como y donde se cargan los ejecutables en memoria fisica.

## Segmentacion y paginacion como solucion:

- Segmentacion: dividew el programa en segmentos logicos (como codigos, datos, pila), permitiendo una gestion mas flexible de la memoria.

- Paginacion: Divide tanto la memoria fisica como la virtual en paginas de tamanio fijo, faciilitando la asignacion y reubicacion de memoria sin preocuparse por la fragmentacion.

## 2. Proteccion entre tareas:

La proteccion entre tareas garantiza que cada proceso o tarea en el SO tenga acceso unicamente a su propio espacio de memoria y recursos asignados, evitando que interfieran o accedan a la memoria de otros procesos. Este aislamiento es crucial para la estabilidad, seguridad y confiabilidad del sistema.


## Segmentacion:

La segmentacion es un metodo de gestion de memoria que divide el espacio de direcciones de un proceso en segmentos logicos de tamanio variable, en lugar de utilizar bloques de tamanio fijo como en la paginacion. Cada segmento representa una entidad logica del progama como:

- Codigo (Text Segment): Contiene las instrucciones ejecutables

- Datos (Data segment): Almacena variaables estaticas y globales

- Pila (Stack Segment): Gestiona las llamadas a funciones y el almacenamiento de variables locales.

- Segmentos adicionales: Pueden incluir segmentos para datos dinamicos, bibliotecas compartidas, etc.

## Como funciona la segmentacion?

## 1. Division Logica:

Cada proceso se divide en multiples segmentos logicos, cada uno con una funcion especifica dentro del programa. Esta division refleja la estructura logica del software, facilitando la gestion y el acceso a diferentes partes del programa de manera independiente.

## 2. Direccionamiento:

El espacio de direcciones virtuales se compone de multiples segmentos, cada uno con su propia direccion base y limite. Una direccion logica en el sistema segmentado se compone de dos partes:

- Numero de Segmento: Identifica a cual de los segmentos pertenece la direccion.

- Desplazamiento (offset): Especifica la ubicacion dentro del segmento

## 3. Traduccion a direcciones fisicas:

El unidad de gestion de memoria (MMU) traduce las direcciones logicas a direcciones fisicas utilizando tablas de segmentos. Cada entrada en la tabla de segmentos contiene:

- Direccion base: Direccion inicial del segmento en la memoria fisica.

- Limite: Tamanio del semgento, que se utiliza para verificar que el acceso este dentro de los limites permitidos.

## Es la segmentacion una solucion?

## 1. Reubicacion de ejecutables:

Si, la segmentacion contribuye a resolver el problema de la reubicacion de ejecutables de las siguientes maneras:

- Flexiblidad en la carga: Al dividiir el programa en segmentos logicos, cada segmento puede cargarse en cualquier ubicacion de la memoria fisica sin afectar a otros segmentoss. Esto facilita la reubicacion dinamica durante la carga o en tiempo de ejecucion.

- Independencia de Segmentos: Los segmentos pueden moverse de manera independiente, lo que reduce la complejidad de la reubicacion comparado con metodos que manejan el programa como una unica unidad contigua.

## 2. Proteccion entre tareas:

Si, la segmentacion tambien aborda la proteccion entre tareas:

- Aislamiento de segmentos: Cada segmento tiene sus propios permisos de acceso (lectura, escritura, ejecucion). Esto asegura que un proceso no pueda acceder o modificar los segmentos de otro proceso, proporcionando un aislamiento efectivo.

- Contol de Acceso: La MMU verifica los permisos durante la traduccion de direcciones, previniendo accesos no autorizados y protegiendo la integridad de cada proceso.
