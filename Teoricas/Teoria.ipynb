Flip Flops

Los flip-flops son dispositivos de memoria que almacenan un bit de informacion. El estado de la salida Q en cualquier momento futuro Q(t + 1) depende de las entradas actuales (S y R) y del estado anterior Q(t).

Flip-Flop SR (Set-Reset) en detalle





El flip-flop SR es uno de los tipos mas basicos de flip-flops y sirve como un dispositivo de almacenamiento que puede mantener un estado binario (0 o 1). A continuacion, se detalla su estructura, funcionamiento, tabla de verdad, diagramas y posibles aplicaciones.

1- Estructura de Flip-Flop SR

El flip-flop SR consta de dos entradas, Set (S) y Reset (R), y dos salidas, Q y Q complemento. La estructura interna tipica se puede construir utilizando compuertas NOR o NAND.

2- Funcionamiento del Flip-Flop SR

El comportamiento del flip-flop SR depende de las entradas S y R, y del estado previo de las salidas Q y Q complemento. Las reglas de funcionamiento se pueden describir en terminos de la tabla de verdad.

3- Estado indeterminado

Cuando ambas entradas S y R son 1 simultaneamente, el estado del flip-flop SR se vuelve indeterminado. Esto puede causar comportamiento no desaeado y es generalmente evitado en aplicaciones practicas.

4- Ejemplo practico

Supongamos que queremos diseniar un sistema que controle una luz. La luz puede estar encedida (Set) o apagada (Reset). Un flip-flop SR puede ser utilizado para recordar el ultimo estado de la liz:

- S = 1, R = 0: La luz se enciende
- S = 0, R = 1: la luz se apaga
- S = 0, R = 0: La luz mantiene su estado actual
- S = 1, R = 1: Estado indeterminado

En resumen, el flip-flop SR es un elemento basico en la construccion de circuitos digitales secuenciales y se utiliza ampliamente en aplicaciones donde es necesario almacenar y mantener un bit de informacion binaria. Evitar la condicion indeterminada y considerar la implementacion de seniales de habilitacion son practicas comunes para asegurar un funcionamiento fiable en aplicaciones practicas.

5- Tabla de verdad del Flip-Flop SR

S	R	Q(t)	Q(t+1)	Descripción
0	0	0	0	Sin cambio
0	0	1	1	Sin cambio
0	1	0	0	Reset (Q se pone en 0)
0	1	1	0	Reset (Q se pone en 0)
1	0	0	1	Set (Q se pone en 1)
1	0	1	1	Set (Q se pone en 1)
1	1	0	Ind	Indeterminado
1	1	1	Ind	Indeterminado

1- Sin cambios (S=0, R=0):
    - Cuando ambas entradas S (Set) y R (Reset) son 0, el estado de la salida Q no cambia
    - Si Q era 0, permanece 0. Si Q era 1, permanece 1.

2- Reset (S=0, R=1):
    - Cuando S es 0 y R es 1, la salida Q se pone en 0, independientemente del estado anterior de Q.
    - Esto significa que el flip-flop se resetea.

3- Set (S=1, R=0):
    - Cuando S es 1 y R es 0, la salida Q se pone en 1, independientemente del estado anterior de Q
    - Esto significa que el flip-flop se setea.

4- Indetermiando (S=1, R=1):
    - Cuando ambas entradas S y R son 1, el estado de la salida Q es indeterminado.
    - Esta condicion es generalmente evitada en el disenio de circuitos porque puede llevar a un comportamiento impredecible.
